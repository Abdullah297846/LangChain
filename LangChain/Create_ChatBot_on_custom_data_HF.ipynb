#Pass the URLs and extract the information

```python
URLs=[
    'https://www.linkedin.com/in/abdullah-nadeem-45b691215/',
    'https://artisticabdullah.netlify.app/',
    'https://www.techtarget.com/searchenterpriseai/definition/BERT-language-model',
    'https://yolov8.com/'

]
```

```python
loaders=UnstructuredURLLoader(urls=URLs)
data=loaders.load()
```

#Split the text into chunks

```python
text_splitter=CharacterTextSplitter(separator='\n',
                                    chunk_size=1000,
                                    chunk_overlap=200)
```

```python
text_chunks=text_splitter.split_documents(data)
```

#Download th Hugging Face Embeddings

```python
embeddings=HuggingFaceEmbeddings()
```

```python
query_result = embeddings.embed_query("Hello world")
len(query_result)
```

#Convert the Text Chunks into Embeddings and Create a Knowledge Base

```python
vectorstore=FAISS.from_documents(text_chunks, embeddings)
```

```python
tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-2-7b-chat-hf",
                                          use_auth_token=True,)


model = AutoModelForCausalLM.from_pretrained("meta-llama/Llama-2-7b-chat-hf",
                                             device_map='auto',
                                             torch_dtype=torch.float16,
                                             use_auth_token=True,
                                              load_in_8bit=True,
                                              #load_in_4bit=True
                                             )
```

```python
pipe = pipeline("text-generation",
                model=model,
                tokenizer= tokenizer,
                torch_dtype=torch.bfloat16,
                device_map="auto",
                max_new_tokens = 512,
                do_sample=True,
                top_k=30,
                num_return_sequences=1,
                eos_token_id=tokenizer.eos_token_id
                )
```

```python
llm=HuggingFacePipeline(pipeline=pipe, model_kwargs={'temperature':0})
```

```python
llm("Kindly tell me about yolov8")
```

#Initialize the Retrieval QA with Source Chain

```python
chain = RetrievalQAWithSourcesChain.from_llm(llm=llm, retriever=vectorstore.as_retriever())
```

```python
result=chain({"question": "Kindly tell me about yolov8?"}, return_only_outputs=True)
```

```python
while True:
  query=input(f"Prompt: ")
  if query == 'exit':
    print('Exiting')
    sys.exit()
  if query =='':
    continue
  result=chain({'question':query})
  print(f"Answer: " +result["answer"])
```
